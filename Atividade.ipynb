  Calcular o módulo
  - [numpy.linalg.norm](https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html)

</details>

```python
# Seu código aqui

def gerar_matriz_distancia(m, n):
    # A função utiliza np.indices para gerar matrizes de coordenadas i e j e aplica a fórmula da distância Euclidiana (raiz quadrada da soma dos quadrados) de forma vetorizada para calcular a distância de cada ponto até a origem [0,0].
    i, j = np.indices((m, n))
    X = np.sqrt(i**2 + j**2)
    return X

# Teste da função com m=5 e n=4
matriz_X = gerar_matriz_distancia(5, 4)
print("Matriz X resultante (distâncias até [0,0]):\n", matriz_X)
```

**b)** Função parabólica dada pela soma do quadrado de suas coordenadas;

```python
# Seu código aqui

# A função np.indices gera as matrizes de coordenadas e o cálculo i**2 + j**2 aplica a soma do quadrado dos índices em cada ponto, resultando em uma superfície parabólica que cresce conforme o afastamento da origem (0,0).
i, j = np.indices((100, 100))
parabolica = i**2 + j**2

plt.imshow(parabolica, cmap='viridis')
plt.colorbar(label='Valor da função')
plt.title("Função Parabólica (i² + j²)")
plt.show()

# Como o valor de cada pixel é o resultado de i² + j², o ponto na coordenada [0,0] terá valor 0 (cor escura). À medida que nos afastamos da origem (aumentando os índices de linha i e coluna j), o valor aumenta de forma quadrática, criando o gradiente característico da parábola que se expande para as extremidades da imagem.
```

**c)** Círculo de raio 40, com centro em (0,0);
